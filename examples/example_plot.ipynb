```python
import numpy as np
from pycop import gaussian

cop = gaussian()

lvls = [0.01,0.1,0.4,0.8,1.3,1.6]

cop.plot_pdf([0.8], plot_type="3d", Nsplit=40, )
cop.plot_pdf([0.8],  plot_type="contour", Nsplit=40,  levels=lvls )
cop.plot_cdf([0.8], plot_type="3d", Nsplit=40,)
lvls = np.linspace(0.1,1,10) 
cop.plot_cdf([0.8], plot_type="contour", Nsplit=40, levels=lvls )
```

Adding Gaussian marginals

```python
from scipy.stats import norm

marginals = [
    {
        "distribution": norm, "loc" : 0, "scale" : 1,
    },
    {
        "distribution": norm, "loc" : 0, "scale": 1,
    }]

cop.plot_mpdf([0.8], marginals, plot_type="3d", Nsplit=50)
cop.plot_mpdf([0.8], marginals, plot_type="contour", Nsplit=50)
```

Clayton Copula

```python
from pycop import archimedean

cop = archimedean(family="rclayton")
cop.plot_pdf([2], plot_type="3d", Nsplit=100, )
cop.plot_cdf([2], plot_type="3d", Nsplit=100,)

lvls = [0.01,0.1,0.4,0.8,1.3,1.6]
cop.plot_pdf([2],  plot_type="contour", Nsplit=100,  levels=lvls )
cop.plot_cdf([2], plot_type="contour", Nsplit=100, levels=np.linspace(0.1,1,10) )
```

Mixture of 2 Copulas

```python
from pycop import mixture

marginals = [
    {
        "distribution": norm, "loc" : 0, "scale" : 1,
    },
    {
        "distribution": norm, "loc" : 0, "scale": 1,
    }]

cop = mixture(["clayton", "gumbel"])

cop.plot_pdf([0.5, 2, 2],  plot_type="contour", Nsplit=40,  levels=[0.01,0.1,0.4,0.8,1.3,1.6] )
cop.plot_mpdf([0.5, 2, 2], marginals, plot_type="contour", Nsplit=50)
```

Mixture of 3 Copulas

```python
cop = mixture(["clayton","gaussian", "gumbel"])

cop.plot_pdf([1/3, 1/3, 1/3, 2, 0.5, 4],  plot_type="contour", Nsplit=40,  levels=[0.01,0.1,0.4,0.8,1.3,1.6] )

cop.plot_mpdf([1/3, 1/3, 1/3, 2, 0.5, 2], marginals, plot_type="contour", Nsplit=50)
```
