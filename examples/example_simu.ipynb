Plot function with Gaussian marginals

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

def sct_plt(u1, u2):
    u1 = norm.ppf(u1)
    u2 = norm.ppf(u2)
    plt.scatter(u1, u2, color="black", alpha=0.6)
    plt.show()
```

Bivariate Gaussian simulations with sample size $m = 1000$ and correlation $\rho =0.8$



```python
from pycop import simulation

n = 2
m = 1000

# We define the correlation matrix as follow:
corrMatrix = np.array([[1, 0.8], [0.8, 1]])
u1, u2 = simulation.simu_gaussian(n, m, corrMatrix)
sct_plt(u1, u2)
```

Student copula simulations with $\nu=2$

```python
corrMatrix = np.array([[1, 0.8], [0.8, 1]])
u1, u2 = simulation.simu_tstudent(n, m, corrMatrix, nu=2)
sct_plt(u1, u2)
```

Clayton copula simulation ($\theta=2$)

```python
familly = "clayton"
u1, u2 = simulation.simu_archimedean(familly, n, m, theta=2)
sct_plt(u1, u2)
```

Simulations in higher dimensions

Gaussian

```python
from pycop import utils

n = 3
m = 1000

# We defined the correlation matrix:
corrMatrix = np.array([[1, 0.9, 0], [0.9, 1, 0], [0, 0, 1]])
u = simulation.simu_gaussian(n, m, corrMatrix)
u = norm.ppf(u)
utils.empiricalplot(u, contour=True)
```

Gumbel simulations

```python
familly = "gumbel"
u = simulation.simu_archimedean(familly, n, m, theta=2)
u = norm.ppf(u)
utils.empiricalplot(u, contour=True)
```
